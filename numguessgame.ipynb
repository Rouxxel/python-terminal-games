```python
import random

print("**********************************")
print("Let's play a number guessing game")
print("**********************************",end="\n\n")

while True:
    
    breakGame=False
    
    print("Please select valid integers for the range and/or a valid range")
    while True:
        minNum=input("Please select a minimum number: ")
        maxNum=input("Please select a maximum number: ")
    
        if (minNum.isdigit() and maxNum.isdigit()): #Because python takes input() as a string
            minNum=int(minNum) #Cast to int
            maxNum=int(maxNum) #Cast to int
        
            if (minNum < maxNum): #Ensure min is less than max
                break 
            else:
                print("Minimum number can't be less or equal to Maximum value")
        else:
            print ("One or both of the given values is not a number")

    print("Range goes ",minNum," to ", maxNum," please guess the number",end="\n\n")
    randomNum= random.randint(minNum,maxNum)

    print("Please select valid integer guess")
    while True:
        guessNum=input("What random number is?: ")
    
        if guessNum.isdigit():
            int(guessNum)
            break
        else:
            print("Please provide a number")

    if(guessNum==randomNum):
        print("Congratulations, your guess ",guessNum," is correct",end="\n\n")
    else:
        print("Sorry, the number was ",randomNum,end="\n\n")
       
        
    print("If you want to play again enter 1, if not enter 2",end="\n\n")
    while True:
        userOption=input("Enter your choice: ")
        
        
        
        if userOption!="1" and userOption!="2":
            print("Invalid option: ",userOption)
        elif userOption=="1":
            print("Thanks for playing")
            breakGame = True
            break
        else:
            print("New game:",end="\n\n")
            break
    
    if breakGame:
        break
```
